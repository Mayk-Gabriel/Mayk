# Exercício 1: Dados Simulados


Análise e Tratamento de valores ausentes e inconstâncias


# 1- Importação das bibliotecas

```python
import pandas as pd
import numpy as np
```

# 2- Criação do dataset simulado com problemas

```python
np.random.seed(123) # para reprodutibilidade
n_rows_ex = 200
dados_ex = {
'peso': np.random.normal(70, 15, n_rows_ex),
'altura': np.random.normal(1.70, 0.15, n_rows_ex),
'idade': np.random.normal(35, 12, n_rows_ex),
'genero': np.random.choice(['M', 'F', 'Masculino', 'Feminino', 'm', 'f'], n_rows_ex),
'atividade_fisica': np.random.choice(['Sedentário', 'Moderado', 'Ativo',
'Muito Ativo', 'sedentario'], n_rows_ex),
'consumo_calorico': np.random.lognormal(7, 0.4, n_rows_ex),
'pressao_arterial_sistolica': np.random.normal(120, 20, n_rows_ex)
}
df_exercicio = pd.DataFrame(dados_ex)
```

# 3- Inserindo problemas de propósito

```python

df_exercicio.loc[df_exercicio.sample(frac=0.1).index, 'peso'] = np.nan
df_exercicio.loc[df_exercicio.sample(frac=0.15).index, 'altura'] = np.nan
df_exercicio.loc[df_exercicio.sample(frac=0.08).index, 'consumo_calorico'] = np.nan

df_exercicio.loc[10, 'peso'] = -70 # peso negativo
df_exercicio.loc[25, 'altura'] = 3.5 # altura irreal
df_exercicio.loc[30, 'idade'] = -5 # idade negativa
```

# 4- Vizualização inicial de dados


```python
df_exercicio.head()
```

# 5 - Análise de Valores ausentes

```python
# Contagem de valores ausentes por coluna
df_exercicio.isna().sum()
```

# 6- Tratamento dos valores ausentes

```python
# Preenchendo valores numéricos ausentes com a média
df_exercicio['peso'].fillna(df_exercicio['peso'].mean(), inplace=True)
df_exercicio['altura'].fillna(df_exercicio['altura'].mean(), inplace=True)
df_exercicio['consumo_calorico'].fillna(df_exercicio['consumo_calorico'].mean(),
inplace=True)
```

# 7- Tratamento das inconsistências

```python
# Detectar valores inconsistentes
df_exercicio[df_exercicio['peso'] < 0]
df_exercicio[df_exercicio['altura'] > 2.5]
df_exercicio[df_exercicio['idade'] < 0]
```

# 8 - Padronização de dados categóricos

```python
# Padronizando gênero
df_exercicio['genero'] = df_exercicio['genero'].replace({
'M': 'Masculino',
'm': 'Masculino',
'F': 'Feminino',
'f': 'Feminino'
})
# Padronizando atividade física
df_exercicio['atividade_fisica'] = df_exercicio['atividade_fisica'].replace({
'sedentario': 'Sedentário'
})
```
